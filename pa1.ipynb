```python
from sqlalchemy import create_engine
from sqlalchemy import MetaData
```

```python
from sqlalchemy import Table
from sqlalchemy import Column
from sqlalchemy import String
from sqlalchemy import Float
from sqlalchemy import Integer
```

```python
import numpy as np  
import pandas as pd 
pd.options.mode.chained_assignment = None
```

```python
df= pd.read_csv("EML-PA1_data.csv", na_values=['NA','NIL'])
df.head()
print(df.info())
```

```python
df_filtered = df[["age",'job','education','duration','success']]
print(df_filtered.info())
df_filtered.head()
```

```python
print("**job value**")
print(df_filtered.job.value_counts())
print("**education**")
print(df_filtered.education.value_counts())
print('**success**')
print(df_filtered.success.value_counts())
print(df_filtered[['age','duration']].describe())
df_filtered.head()
```

```python
df_filtered.job.replace({'unknown':np.NaN,"admin.":"admin"},inplace=True)
df_filtered.job.value_counts(dropna=False)

df_filtered.head()
```

```python
df_filtered.success.replace({'yes':'y','no':'n'},inplace=True)
df_filtered.success.value_counts(dropna=False)
df_filtered.info()
df_filtered.head()
```

```python
df_filtered['duration']=df_filtered['duration'].apply(pd.to_numeric,errors='coerce')
print(df_filtered.info())
df_filtered.head()
```

```python
df_filtered.loc[df_filtered["age"].between(25,39),"age_cat"]=0
df_filtered.loc[df_filtered["age"].between(40,54),"age_cat"]=1
df_filtered.loc[df_filtered["age"].between(55,70),"age_cat"]=2
df_filtered["age"]=df_filtered["age"].astype('category')
df_filtered["age_cat"].value_counts()
```

```python
df_filtered.education.value_counts()
```

```python
new_education={"education":{"unknown":0,"primary":1,"secondary":2,"tertiary":3}}
df_filtered.replace(new_education,inplace=True)
print(df_filtered.head())
```

```python
print(df_filtered["education"].value_counts())
print(df_filtered.info())
print((df_filtered.head()))
df_filtered.head()
```

```python
df_filtered.dropna(inplace=True)
df_filtered.info()
df_filtered.head()
```

```python
engine=create_engine('sqlite:///pa1.sqlite')
```

```python
df_filtered= df_filtered[['age','job','education','duration','success','age_cat']]
```

```python
con = engine.connect()
df_filtered.to_sql("Pa1",con,if_exists="replace")
```

```python
metadata=MetaData()
```

```python
PA1=Table('Pa1',metadata,
         Column('age',Integer()),
         Column('job',String(255)),
         Column('education',Integer()),
         Column('duration',Float()),
         Column('success',Integer()),
         Column('age_cat',Integer()),
         )
metadata.create_all(engine)
```

```python
names = ['age','education','duration','success']
```

```python
df_filtered.colums = names
df_filtered.head()
```

```python
result = pd.read_sql_query("SELECT * FROM Pa1 limit 20",con)
result
```

```python
result = pd.read_sql_query("select Count(*) AS total from Pa1",con)
result
```

```python
result = pd.read_sql_query("Select Count(success) As total from Pa1 where success='y' ",con)
result
```

```python
result = pd.read_sql_query("select COUNT(age_cat) as profile from Pa1 where age BETWEEN 40 AND 54",con)
result
```
